```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import matplotlib as mpl
import seaborn as sns
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
data1 = pd.read_csv('ghaxfak line.csv')
data1.head()
```

```python
fak_plot =  data1['fak']
gammah2ax_plot =  data1['gh2ax']

distance_microns = data1['distancia']

fak_plot_normalized = fak_plot/max(fak_plot)
gammah2ax_plot_normalized = gammah2ax_plot/max(gammah2ax_plot)
```

```python
Nome_imagem_pdf = 'plot_fak-h2ax_loci-gammah2ax_withxy_norm_set22_v2.pdf'
Nome_imagem_jpeg = 'plot_fak-h2ax_loci-gammah2ax_withxy_norm_set22_v2.jpeg'
# Sequences
x_axes = distance_microns
line1 = fak_plot_normalized
line3 = gammah2ax_plot_normalized

# infor para mudar fcil

Tamanho_ticks_Y = 16
Tamanho_label_Y = 16
Eixo_Y = "Normalized intensity"
Eixo_X = "Distance (m)"

fig, ax = plt.subplots(dpi=200)

# Plot linear sequence, and set tick labels to the same color
ax.plot(x_axes, line1, color='green')

#parametros do eixo y
ax.tick_params(axis='y', labelcolor='black')
plt.yticks((0.0,0.5,1.0),fontsize=Tamanho_ticks_Y)
ax.set_ylabel(Eixo_Y,fontsize=Tamanho_label_Y)

#parametros do eixo y
plt.xticks(fontsize=Tamanho_ticks_Y)
ax.set_xlabel(Eixo_X,fontsize=Tamanho_label_Y)


# Generate a new Axes instance, on the twin-X axes (same position)
# ax2 = ax.twinx()

# Plot exponential sequence, set scale to logarithmic and change tick color
# ax2.plot(x_axes, line2, color='red')
#ax2.tick_params(axis='y', labelcolor='red')

# Generate a new Axes instance, on the twin-X axes (same position)
ax3 = ax.twinx()

# Plot exponential sequence, set scale to logarithmic and change tick color
ax3.plot(x_axes, line3, color='blue')
#ax3.tick_params(axis='y', labelcolor='blue')

# Hide the y axes
#ax.axes.yaxis.set_visible(False)
# ax2.axes.yaxis.set_visible(False)
ax3.axes.yaxis.set_visible(False)

# Hide the right and top spines
sns.despine()

# Hide the x axes
#ax.axes.xaxis.set_visible(False)

point1 = [.36, .3]
point2 = [.64, .3]

x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]


def traco(text,X,Y):
    x = (X[0]+X[1])/2
    y = 1.2* max(Y[0], Y[1]) #caso o trao esteja no lugar errado,  aqui onde arruma, no valor antes da multiplicao
    yp = y-.2
    xp = x-.01 # caso voc achar que o astersco no esta no lugar certo, altere o valor que  subtraido do x at que ele aparea onde  desejado
    
#     props = {'connectionstyle':'bar', ,'arrowstyle':'-',
#             'shrinkA':0,'shrinkB':0, 'linewidth':4,'color':"black"}
    props = dict(connectionstyle= "bar,fraction=-0.3", arrowstyle='-',
            shrinkA=0, shrinkB=0, linewidth=4, color="black")
    
    ax.annotate(text, fontsize =20, xy=(xp,yp-0.1*max(Y)), zorder=10, color="#000000")
    ax.annotate('', xy=(X[0],y), xytext=(X[1],y), arrowprops=props)
    
traco('~0.28m', x_values, y_values )

plt.savefig(Nome_imagem_pdf,bbox_inches='tight')
plt.savefig(Nome_imagem_jpeg,bbox_inches='tight')
plt.show()
```
