```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import matplotlib as mpl
import seaborn as sns
from scipy.stats import ttest_ind
from scipy.stats import sem
%matplotlib inline
```

```python
# dados de agosto de 2022, provavelmente a terceira vez que foi feita com o Pedro (segunda dele)
inibi_data = pd.read_csv('data/inibicao fak limpo.csv')
inibi_data.head()
```

```python
filtro_ct_dmso = (inibi_data['tto'] == 'ct_dmso')
filtro_dox_dmso = (inibi_data['tto'] == 'dox_dmso')
filtro_dox_pf = (inibi_data['tto'] == 'dox_pf')
```

```python
#--------------------------------------------------
data = inibi_data
#--------------------------------------------------
tto1 = 'ct_dmso'

filtro = (data['tto'] == tto1)
fak1 = data.loc[filtro, 'ctcf_fak']
ku801 = data.loc[filtro, 'ctcf_ku80']
gh2ax1 = data.loc[filtro, 'ctcf_gh2ax']

# dados fak
fak_med1 = fak1.mean()
fak_sem1 = fak1.sem()

# dados ku80
ku80_med1 = ku801.mean()
ku80_sem1 = ku801.sem()

# dados gh2ax
gh2ax_med1 = gh2ax1.mean()
gh2ax_sem1 = gh2ax1.sem()

#--------------------------------------------------
tto2 = 'dox_dmso'

filtro = (data['tto'] == tto2)
fak2 = data.loc[filtro, 'ctcf_fak']
ku802 = data.loc[filtro, 'ctcf_ku80']
gh2ax2 = data.loc[filtro, 'ctcf_gh2ax']

# dados fak
fak_med2 = fak2.mean()
fak_sem2 = fak2.sem()

# dados ku80
ku80_med2 = ku802.mean()
ku80_sem2 = ku802.sem()

# dados gh2ax
gh2ax_med2 = gh2ax2.mean()
gh2ax_sem2 = gh2ax2.sem()
#--------------------------------------------------
tto3 = 'dox_pf'

filtro = (data['tto'] == tto3)
fak3 = data.loc[filtro, 'ctcf_fak']
ku803 = data.loc[filtro, 'ctcf_ku80']
gh2ax3 = data.loc[filtro, 'ctcf_gh2ax']

# dados fak
fak_med3 = fak3.mean()
fak_sem3 = fak3.sem()

# dados ku80
ku80_med3 = ku803.mean()
ku80_sem3 = ku803.sem()

# dados gh3ax
gh2ax_med3 = gh2ax3.mean()
gh2ax_sem3 = gh2ax3.sem()
#--------------------------------------------------
fak_med_list = [fak_med1,fak_med2 ,fak_med3]
fak_sem_list = [fak_sem1,fak_sem2 , fak_sem3]
fak_list = [fak1 ,fak2 ,fak3 ]

ku80_med_list = [ku80_med1,ku80_med2 ,ku80_med3]
ku80_sem_list = [ku80_sem1,ku80_sem2 , ku80_sem3]
ku80_list = [ku801 ,ku802 ,ku803 ]

gh2ax_med_list = [gh2ax_med1,gh2ax_med2 ,gh2ax_med3]
gh2ax_sem_list = [gh2ax_sem1,gh2ax_sem2 , gh2ax_sem3]
gh2ax_list = [gh2ax1 ,gh2ax2 ,gh2ax3 ]
```

```python
# Dados ######################################### ct dmso x dox dmso (fak)

ttest_group = [fak_list[0], fak_list[1]]#grupo onde tem todos os dados que sero calculados o ttest
means = [fak_med_list[0], fak_med_list[1]] #grupo onde tem as informaes da mdias
erro = [fak_sem_list[0], fak_sem_list[1]] #grupo que contem a informao dos erros

cor_ct = 'white'              #cor do grupo controle (barra1)
cor_dox = 'black'              #cor do grupo dox (barra2)

Nome_imagem_pdf = 'inibi_fak_CTCF_fak_ctdmso-doxdmso_ago2022.pdf' # nome da imagem salva em .pdf
Nome_imagem_jpeg = 'inibi_fak_CTCF_fak_ctdmso-doxdmso_ago2022.jpeg' # nome da imagem salva em .jpeg

cor_envolta = "black" #cor da linha envolta das barras
espessura_volta = 5   #espessura da linha envolta das barras
tamanho_barra_erro = 30           #tamanho da barra horizontal de erro

# coisas relacionadas ao eixo Y
Eixo_Y = 'Nucleus FAK CTCF' # O que vai aparecer escrito no eixo Y
Tamanho_label_Y = 46 # Tamanho do texto do eixo Y
Tamanho_ticks_Y = 30 # Tamanho dos nmeros que aparecem no eixo Y

# coisas relacionadas ao eixo X
labels = ['DMSO', 'Dox + DMSO'] #o que vai aparecer embaixo das barras, no eixo X.
Tamanho_label_X = 46 # Tamanho do texto do eixo X

print(means)

# normalizao ######################################### 
# essa parte serve para criar um grfico que visivelmente fica mais fcil ver quantas vezes ficou maior o grupo tratado (barra 2) do controle (barra 1)
# caso no quiser isso,  s comentar todas essa parte ou apagar

erro_ct = erro[0]/means[0]
erro_dox = erro[1]/means[0]
erro = [erro_ct,erro_dox]

ct_norm = means[0]/means[0]
dox_norm = means[1]/means[0]
means = [ct_norm, dox_norm]

# p #########################################
# essa parte  uma funo que pega as datas dos dois grupos e faz um teste estatstico t de studant para ver se  uma variao significativa ou no e quanto significativa.
# aparece em asterstico automaticamente no grfico

stat, p = ttest_ind(ttest_group[0], ttest_group[1])
print('t=%.3f, p=%.5f' % (stat, p))

def convert_pvalue_to_asterisks(pvalue):
    if pvalue <= 0.0001:
        return "****"
    elif pvalue <= 0.001:
        return "***"
    elif pvalue <= 0.01:
        return "**"
    elif pvalue <= 0.05:
        return "*"
    return "NS"

pValue = convert_pvalue_to_asterisks(p)
print(convert_pvalue_to_asterisks(p))

# plotar trao #########################################
# essa parte serve para criar aquela barra encima do das duas barras
# agradecimentos para Sofia que me mostrou isso pela primeira vez e onde achou
# agradecimento ao Bruno que  o melhor programador que conheo e me ajudou.


def traco(text,X,Y):
    x = (X[0]+X[1])/2
    y = 1.2* max(Y[0], Y[1]) #caso o trao esteja no lugar errado,  aqui onde arruma, no valor antes da multiplicao
    dx = abs(X[0]-X[1])
    yp = y 
    xp = x-.005 # caso voc achar que o astersco no esta no lugar certo, altere o valor que  subtraido do x at que ele aparea onde  desejado
       
    props = {'connectionstyle':'bar','arrowstyle':'-',\
            'shrinkA':20,'shrinkB':20,'linewidth':4,'color':"black"}
    ax.annotate(text,fontsize =46, xy=(xp,yp+0.2*max(means)), zorder=10,color="#000000")
    ax.annotate('', xy=(X[0],y), xytext=(X[1],y), arrowprops=props)

# grfico #########################################
x = [.045,.075]  # localizao da barra
width = 0.02# espessura da barra

fig, ax = plt.subplots(figsize=(6,9), dpi=80)
barra_controle = ax.bar(x[0], means[0], yerr=erro[0], capsize= tamanho_barra_erro, width= width, color= cor_ct ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[0])
barra_doxo =  ax.bar(x[1], means[1], yerr=erro[1], capsize= tamanho_barra_erro, width= width ,color= cor_dox ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[1])
traco(pValue,x,means)
#rects = ax.patches

# Os nmeros em cima das barras #########################################
#label = ["%.2f"%item for item in means] # essa parte de "%.2f"%  para aparecer o numero em float, se trocar pra i,  inteiro
#for rect, label in zip(rects, label):
    #height = rect.get_height()
    #ax.text(rect.get_x() + rect.get_width() / 2, height+0.2*max(means), label,
    #        ha='center', va='bottom')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel(Eixo_Y,fontsize=Tamanho_label_Y)
#ax.set_title('mdia de FAK no ncleo - 24h doxo',fontsize=16, pad = 20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=Tamanho_label_X, rotation=45, ha='right')
# plt.ylim(0, 1.6* max(means))
plt.ylim(0,1.5* max(means))
plt.xlim((x[0]-.02),(x[1])+.02)
#plt.legend(prop={"size":15},loc="upper left")
plt.yticks((0,0.5,1,1.5,2),fontsize=Tamanho_ticks_Y)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.savefig(Nome_imagem_pdf,bbox_inches='tight') # a parte do bbox_inches='tight'  que faz o tamanho ficar correto
plt.savefig(Nome_imagem_jpeg,bbox_inches='tight')

plt.show()
```

```python


# dados FAK

grupos = [fak1, fak2, fak3]
means = [fak_med1, fak_med2, fak_med3]
erros = [fak1.std(), fak2.std(), fak3.std()]

print (means)

nome_imagem = 'inifak fak ctcf 3 groups ago2022 v5'
Nome_imagem_pdf = nome_imagem + '.pdf' # nome da imagem salva em .pdf
Nome_imagem_jpeg = nome_imagem + '.jpeg' # nome da imagem salva em .jpeg

cores_barras = ['white', 'red', 'brown']
cor_envolta = "black" #cor da linha envolta das barras
espessura_volta = 4   #espessura da linha envolta das barras
tamanho_barra_erro = 30           #tamanho da barra horizontal de erro

# coisas relacionadas ao eixo Y
Eixo_Y = 'Nucleus FAK CTCF' # O que vai aparecer escrito no eixo Y
Tamanho_label_Y = 46 # Tamanho do texto do eixo Y
Tamanho_ticks_Y = 30 # Tamanho dos nmeros que aparecem no eixo Y

# coisas relacionadas ao eixo X
labels = ['DMSO', 'Dox + DMSO', 'Dox + PF'] #o que vai aparecer embaixo das barras, no eixo X.
Tamanho_label_X = 46 # Tamanho do texto do eixo X

means_normalizado = []
erros_normalizado = []

for mean in means:
    means_normalizado.append(mean/means[0])

for erro in erros:
    erros_normalizado.append(erro/means[0])

def ttesting(a,b):
    stat, p = ttest_ind(a, b)
    print('t=%.3f, p=%.5f' % (stat, p))

    def convert_pvalue_to_asterisks(pvalue):
        if pvalue <= 0.0001:
            return "****"
        elif pvalue <= 0.001:
            return "***"
        elif pvalue <= 0.01:
            return "**"
        elif pvalue <= 0.05:
            return "*"
        return "NS"

    pValue = convert_pvalue_to_asterisks(p)
    print(convert_pvalue_to_asterisks(p))
    return pValue 
    
    
# plotar trao #########################################
# essa parte serve para criar aquela barra encima do das duas barras
# agradecimentos para Sofia que me mostrou isso pela primeira vez e onde achou
# agradecimento ao Bruno que  o melhor programador que conheo e me ajudou.


def traco(text,X,Y,hight, texthight):
    x = (X[0]+X[1])/2
    y = 1.4* max(Y[0], Y[1])+hight #caso o trao esteja no lugar errado,  aqui onde arruma, no valor antes da multiplicao
    dx = abs(X[0]-X[1])
    yp = y 
    xp = x-.3 # caso voc achar que o astersco no esta no lugar certo, altere o valor que  subtraido do x at que ele aparea onde  desejado
       
    props = {'connectionstyle':'bar','arrowstyle':'-',\
            'shrinkA':20,'shrinkB':20,'linewidth':4,'color':"black"}
    ax.annotate(text,fontsize =46, xy=(xp,yp+0.15*max(means_normalizado)+texthight), zorder=10,color="#000000")
    ax.annotate('', xy=(X[0],y), xytext=(X[1],y), arrowprops=props)

# grfico #########################################   
    
x = [1, 2, 3]  # localizao da barra
width = 0.7# espessura da barra

fig, ax = plt.subplots(figsize=(6,8), dpi=80)
barra_1 = ax.bar(x[0], means_normalizado[0],yerr=erros_normalizado[0],  capsize =28, width= width, color= cores_barras[0] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[0])
barra_2 = ax.bar(x[1], means_normalizado[1],yerr=erros_normalizado[1],  capsize =28, width= width, color= cores_barras[1] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[1])
barra_3 = ax.bar(x[2], means_normalizado[2],yerr=erros_normalizado[2],  capsize =28, width= width, color= cores_barras[2] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[2])
# barra_4 = ax.bar(x[3], means_normalizado[3],yerr=erros_normalizado[3],  capsize =28, width= width, color= cores_barras[3] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[3])

# chamando def para os traos

traco(ttesting(grupos[0],grupos[1]),x[0:2],means_normalizado[0:2],0,0.07)
traco(ttesting(grupos[1],grupos[2]),x[1:3],means_normalizado[1:3],.15,0.07)
traco(ttesting(grupos[0],grupos[2]), (x[0], x[2]), (means_normalizado[0], means_normalizado[2]), 0.3, .25)

plt.ylim(0,2.4* max(means_normalizado))
ax.set_ylabel(Eixo_Y,fontsize=Tamanho_label_Y)
plt.yticks(fontsize=Tamanho_ticks_Y)

ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=Tamanho_label_X, rotation=45, ha='right')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig(Nome_imagem_pdf,bbox_inches='tight') # a parte do bbox_inches='tight'  que faz o tamanho ficar correto
plt.savefig(Nome_imagem_jpeg,bbox_inches='tight')

plt.show()
```
