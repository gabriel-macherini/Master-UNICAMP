```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import matplotlib as mpl
import seaborn as sns
from scipy.stats import ttest_ind
from scipy.stats import sem
%matplotlib inline
```

```python
inibi_data = pd.read_csv('data/resultado final.csv')
inibi_data.tail()
```

```python
def tweak_clean(inibi_data):
    return (inibi_data
     .drop(columns=['index', 'index_original', 'Column1', 'Column2', 'Column3'])
     .rename(columns=lambda col_name: col_name.replace(' ', '_'))
    )

inibi_clean = tweak_clean(inibi_data)
```

```python
filtro_dox_dmso = (inibi_clean['tto'] == 'Dox 12 DMSO 12')
filtro_dox_pf = (inibi_data['tto'] == 'Dox 12 PF 12')

#--------------------------------------------------
data = inibi_clean

tto2 = 'Dox 12 DMSO 12'

filtro = (data['tto'] == tto2)
fak2 = data.loc[filtro, 'fak_CTCF']
ku802 = data.loc[filtro, 'ku80_CTCF']
gh2ax2 = data.loc[filtro, 'gh2ax_CTCF']

# dados fak
fak_med2 = fak2.mean()
fak_sem2 = fak2.sem()

# dados ku80
ku80_med2 = ku802.mean()
ku80_sem2 = ku802.sem()

# dados gh2ax
gh2ax_med2 = gh2ax2.mean()
gh2ax_sem2 = gh2ax2.sem()

#--------------------------------------------------
tto3 = 'Dox 12 PF 12'

filtro = (data['tto'] == tto3)
fak3 = data.loc[filtro, 'fak_CTCF']
ku803 = data.loc[filtro, 'ku80_CTCF']
gh2ax3 = data.loc[filtro, 'gh2ax_CTCF']

# dados fak
fak_med3 = fak3.mean()
fak_sem3 = fak3.sem()

# dados ku80
ku80_med3 = ku803.mean()
ku80_sem3 = ku803.sem()

# dados gh3ax
gh2ax_med3 = gh2ax3.mean()
gh2ax_sem3 = gh2ax3.sem()


#--------------------------------------------------

fak_med_list = [fak_med2 ,fak_med3]
fak_sem_list = [fak_sem2 , fak_sem3]
fak_list = [fak2 ,fak3 ]

ku80_med_list = [ku80_med2 ,ku80_med3]
ku80_sem_list = [ku80_sem2 , ku80_sem3]
ku80_list = [ku802 ,ku803 ]

gh2ax_med_list = [gh2ax_med2 ,gh2ax_med3]
gh2ax_sem_list = [gh2ax_sem2 , gh2ax_sem3]
gh2ax_list = [gh2ax2 ,gh2ax3 ]
```

```python
inibi_clean['tto'].value_counts()
```

```python
# Dados ######################################### dox dmso x dox pf (fak)

ttest_group = [fak_list[0], fak_list[1]]#grupo onde tem todos os dados que sero calculados o ttest
means = [fak_med_list[0], fak_med_list[1]] #grupo onde tem as informaes da mdias
erro = [fak_sem_list[0], fak_sem_list[1]] #grupo que contem a informao dos erros

cor_ct = 'white'              #cor do grupo controle (barra1)
cor_dox = 'black'              #cor do grupo dox (barra2)

Nome_imagem_pdf = 'inifak_CTCF_fak_doxdmso-doxpf.pdf' # nome da imagem salva em .pdf
Nome_imagem_jpeg = 'inifak_CTCF_fak_doxdmso-doxpf.jpeg' # nome da imagem salva em .jpeg

cor_envolta = "black" #cor da linha envolta das barras
espessura_volta = 5   #espessura da linha envolta das barras
tamanho_barra_erro = 30           #tamanho da barra horizontal de erro

# coisas relacionadas ao eixo Y
Eixo_Y = 'Nucleus FAK CTCF' # O que vai aparecer escrito no eixo Y
Tamanho_label_Y = 46 # Tamanho do texto do eixo Y
Tamanho_ticks_Y = 30 # Tamanho dos nmeros que aparecem no eixo Y

# coisas relacionadas ao eixo X
labels = ['Dox + DMSO', 'Dox + PF'] #o que vai aparecer embaixo das barras, no eixo X.
Tamanho_label_X = 46 # Tamanho do texto do eixo X

print(means)

# normalizao ######################################### 
# essa parte serve para criar um grfico que visivelmente fica mais fcil ver quantas vezes ficou maior o grupo tratado (barra 2) do controle (barra 1)
# caso no quiser isso,  s comentar todas essa parte ou apagar

erro_ct = erro[0]/means[0]
erro_dox = erro[1]/means[0]
erro = [erro_ct,erro_dox]

ct_norm = means[0]/means[0]
dox_norm = means[1]/means[0]
means = [ct_norm, dox_norm]

# p #########################################
# essa parte  uma funo que pega as datas dos dois grupos e faz um teste estatstico t de studant para ver se  uma variao significativa ou no e quanto significativa.
# aparece em asterstico automaticamente no grfico

stat, p = ttest_ind(ttest_group[0], ttest_group[1])
print('t=%.3f, p=%.5f' % (stat, p))

def convert_pvalue_to_asterisks(pvalue):
    if pvalue <= 0.0001:
        return "****"
    elif pvalue <= 0.001:
        return "***"
    elif pvalue <= 0.01:
        return "**"
    elif pvalue <= 0.05:
        return "*"
    return "NS"

pValue = convert_pvalue_to_asterisks(p)
print(convert_pvalue_to_asterisks(p))

# plotar trao #########################################
# essa parte serve para criar aquela barra encima do das duas barras
# agradecimentos para Sofia que me mostrou isso pela primeira vez e onde achou
# agradecimento ao Bruno que  o melhor programador que conheo e me ajudou.


def traco(text,X,Y):
    x = (X[0]+X[1])/2
    y = 1.2* max(Y[0], Y[1]) #caso o trao esteja no lugar errado,  aqui onde arruma, no valor antes da multiplicao
    dx = abs(X[0]-X[1])
    yp = y 
    xp = x-.005 # caso voc achar que o astersco no esta no lugar certo, altere o valor que  subtraido do x at que ele aparea onde  desejado
       
    props = {'connectionstyle':'bar','arrowstyle':'-',\
            'shrinkA':20,'shrinkB':20,'linewidth':4,'color':"black"}
    ax.annotate(text,fontsize =46, xy=(xp,yp+0.2*max(means)), zorder=10,color="#000000")
    ax.annotate('', xy=(X[0],y), xytext=(X[1],y), arrowprops=props)

# grfico #########################################
x = [.045,.075]  # localizao da barra
width = 0.02# espessura da barra

fig, ax = plt.subplots(figsize=(6,9), dpi=80)
barra_controle = ax.bar(x[0], means[0], yerr=erro[0], capsize= tamanho_barra_erro, width= width, color= cor_ct ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[0])
barra_doxo =  ax.bar(x[1], means[1], yerr=erro[1], capsize= tamanho_barra_erro, width= width ,color= cor_dox ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[1])
traco(pValue,x,means)
#rects = ax.patches

# Os nmeros em cima das barras #########################################
#label = ["%.2f"%item for item in means] # essa parte de "%.2f"%  para aparecer o numero em float, se trocar pra i,  inteiro
#for rect, label in zip(rects, label):
    #height = rect.get_height()
    #ax.text(rect.get_x() + rect.get_width() / 2, height+0.2*max(means), label,
    #        ha='center', va='bottom')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel(Eixo_Y,fontsize=Tamanho_label_Y)
#ax.set_title('mdia de FAK no ncleo - 24h doxo',fontsize=16, pad = 20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=Tamanho_label_X, rotation=45, ha='right')
# plt.ylim(0, 1.6* max(means))
plt.ylim(0,1.5* max(means))
plt.xlim((x[0]-.02),(x[1])+.02)
#plt.legend(prop={"size":15},loc="upper left")
plt.yticks((0,0.5,1,1.5,2),fontsize=Tamanho_ticks_Y)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.savefig(Nome_imagem_pdf,bbox_inches='tight') # a parte do bbox_inches='tight'  que faz o tamanho ficar correto
plt.savefig(Nome_imagem_jpeg,bbox_inches='tight')

plt.show()
```
