```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import matplotlib as mpl
import seaborn as sns
from scipy.stats import ttest_ind
import re
%matplotlib inline
```

```python
# dados TODOS os rois
ku80_allcells = pd.read_csv("data/pc_celulas.csv")
# ku80_allcells = pd.read_csv("data/pc_cells.csv")

# ----------------------------------------------------------------------------------------------------------------------------------
def coef_pc(data,nome_data):
    '''Calcula coeficiente de pearson das colunas id_first e id_second da data selecionada com o nome da data'''
    first_array = data['id_gh2ax'] 
    second_array = data['id_fak'] 
    pearson_first_second = np.corrcoef(first_array, second_array)[0][1]
    print("O coeficiente de pearson do: " + str(nome_data) + " é " + str(pearson_first_second))
    string_para_print = ("Correlação de pearson: \n" + str(pearson_first_second))
    return string_para_print
# ----------------------------------------------------------------------------------------------------------------------------------
```

# Explicação

Pega os dados de correlação de pearson das imagens dos gh2ax loci dos tratamentos onde cada loci é tratado como um pixel e a correlação é dada por célula.

em detalhes: 

* É extraído cada gh2ax com uma margem extra de 0.5 micron
* É calculado a intensidade integrada de cada um desses rois dos canais de cada proteína
* Cada um desses pontos é agregado por célula e feito uma correlação de pearson
* É salvo em um .csv para feita a imagem aqui.

Obs: Células com menos de 3 rois não são consideradas pois não representam a quantidade de dano que estudamos. Dessa tentativa
