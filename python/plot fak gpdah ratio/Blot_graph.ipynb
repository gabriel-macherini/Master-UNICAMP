```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import matplotlib as mpl
import seaborn as sns

%matplotlib inline
```

```python
# dados extraidos das imagens de blot e quantificadas no imagej

fak_gpadh = [.16, .14, .12, .06]
grupos = fak_gpadh
grupos_normalizado = []

for grupo in grupos:
    grupos_normalizado.append(grupo/grupos[0])

nome_imagem = 'sifak wb bandas'
Nome_imagem_pdf = nome_imagem + '.pdf' # nome da imagem salva em .pdf
Nome_imagem_jpeg = nome_imagem + '.jpeg' # nome da imagem salva em .jpeg

cores_barras = ['white', 'white', 'black', 'black']
cor_envolta = ["black", 'white'] #cor da linha envolta das barras
espessura_volta = 4   #espessura da linha envolta das barras
tamanho_barra_erro = 30           #tamanho da barra horizontal de erro

# coisas relacionadas ao eixo Y
Eixo_Y = 'FAK/GAPDH' # O que vai aparecer escrito no eixo Y
Tamanho_label_Y = 46 # Tamanho do texto do eixo Y
Tamanho_ticks_Y = 30 # Tamanho dos nmeros que aparecem no eixo Y

# coisas relacionadas ao eixo X
labels = ['SiCT', 'SiFAK', 'Dox + SiCT', 'Dox + SiFAK'] #o que vai aparecer embaixo das barras, no eixo X.
Tamanho_label_X = 46 # Tamanho do texto do eixo X

plt.rcParams['hatch.linewidth'] = 3 # regula a espessura das linhas dentro das barras

x = [1, 2, 3, 4]  # localizao da barra
width = 0.7# espessura da barra

fig, ax = plt.subplots(figsize=(6,8), dpi=80)
barra_1 = ax.bar(x[0], grupos_normalizado[0], width= width, color= cores_barras[0] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[0])
barra_2 = ax.bar(x[1], grupos_normalizado[1], width= width, color= cores_barras[1] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[1])
barra_3 = ax.bar(x[2], grupos_normalizado[2], width= width, color= cores_barras[2] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[2])
barra_4 = ax.bar(x[3], grupos_normalizado[3], width= width, color= cores_barras[3] ,edgecolor= cor_envolta, linewidth= espessura_volta, label= labels[3])

barra_1 = ax.bar(x[0], grupos_normalizado[0], width= width, color= 'None' , hatch='', edgecolor= cor_envolta, linewidth= 0, label= labels[0])
barra_2 = ax.bar(x[1], grupos_normalizado[1], width= width, color= 'None' , hatch='/',edgecolor= cor_envolta, linewidth= 0, label= labels[1])
barra_3 = ax.bar(x[2], grupos_normalizado[2], width= width, color= 'None' , hatch='',edgecolor= cor_envolta[1], linewidth= 0, label= labels[2])
barra_4 = ax.bar(x[3], grupos_normalizado[3], width= width, color= 'None' , hatch='/',edgecolor= cor_envolta[1], linewidth= 0, label= labels[3])

ax.set_ylabel(Eixo_Y,fontsize=Tamanho_label_Y)
plt.yticks(fontsize=Tamanho_ticks_Y)

ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=Tamanho_label_X, rotation=45, ha='right')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig(Nome_imagem_pdf,bbox_inches='tight') # a parte do bbox_inches='tight'  que faz o tamanho ficar correto
plt.savefig(Nome_imagem_jpeg,bbox_inches='tight')

plt.show()
```
