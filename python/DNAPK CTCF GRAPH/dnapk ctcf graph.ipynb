```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import matplotlib as mpl
import seaborn as sns
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
data1 = pd.read_csv('data/resultados_dnapk.csv')
data1.head()
```

```python
doxo = data1['tto'] == 'doxo'
controle = data1['tto'] == 'controle'
```

```python
# fak codigo escalavel
# grupos 
fak_ctcf_controle = data1.loc[controle, 'ctcf_fak']
fak_ctcf_doxo = data1.loc[doxo, 'ctcf_fak']

#media
fak_ctcf_controle_med = fak_ctcf_controle.mean()
fak_ctcf_doxo_med = fak_ctcf_doxo.mean()

#std
fak_ctcf_controle_std = fak_ctcf_controle.std()
fak_ctcf_doxo_std = fak_ctcf_doxo.std()

#erro do std
fak_ctcf_controle_err = fak_ctcf_controle.std() / np.sqrt(len(fak_ctcf_controle))
fak_ctcf_doxo_err = fak_ctcf_doxo.std() / np.sqrt(len(fak_ctcf_doxo))

#grupos em listas
fak_ctcf_list = [fak_ctcf_controle,fak_ctcf_doxo]
fak_ctcf_med_list = [fak_ctcf_controle_med,fak_ctcf_doxo_med]
fak_ctcf_std_list = [fak_ctcf_controle_std,fak_ctcf_doxo_std]
fak_ctcf_err_list = [fak_ctcf_controle_err,fak_ctcf_doxo_err]

stat, p = ttest_ind(fak_ctcf_controle, fak_ctcf_doxo)
print('fak_controle/doxo : t=%.3f, p=%.5f' % (stat, p))
```

```python
# dnapk codigo escalavel
# grupos 
dnapk_ctcf_controle = data1.loc[controle, 'ctcf_dnapk']
dnapk_ctcf_doxo = data1.loc[doxo, 'ctcf_dnapk']

#media
dnapk_ctcf_controle_med = dnapk_ctcf_controle.mean()
dnapk_ctcf_doxo_med = dnapk_ctcf_doxo.mean()

#std
dnapk_ctcf_controle_std = dnapk_ctcf_controle.std()
dnapk_ctcf_doxo_std = dnapk_ctcf_doxo.std()

#erro do std
dnapk_ctcf_controle_err = dnapk_ctcf_controle.std() / np.sqrt(len(dnapk_ctcf_controle))
dnapk_ctcf_doxo_err = dnapk_ctcf_doxo.std() / np.sqrt(len(dnapk_ctcf_doxo))

#grupos em listas
dnapk_ctcf_list = [dnapk_ctcf_controle,dnapk_ctcf_doxo]
dnapk_ctcf_med_list = [dnapk_ctcf_controle_med,dnapk_ctcf_doxo_med]
dnapk_ctcf_std_list = [dnapk_ctcf_controle_std,dnapk_ctcf_doxo_std]
dnapk_ctcf_err_list = [dnapk_ctcf_controle_err,dnapk_ctcf_doxo_err]

stat, p = ttest_ind(dnapk_ctcf_controle, dnapk_ctcf_doxo)
print('dnapk_controle/doxo : t=%.3f, p=%.5f' % (stat, p))
```
